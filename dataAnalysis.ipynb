Developers Education and Developing skills

```python
# Learn to code
#learn_2_code_df = df[["ResponseId", "EdLevel","DevType","MainBranch","LearnCode","LearnCodeOnline","LearnCodeCoursesCert"]]
edu_df = df[["ResponseId", "EdLevel","DevType","MainBranch","Gender","LearnCode","LearnCodeOnline","LearnCodeCoursesCert"]]

# Clean dataset
edu_df.isna().sum()

edu_df['EdLevel'].replace(np.nan,'No answer', regex=True, inplace=True)
edu_df['DevType'].replace(np.nan,'No answer',regex=True,inplace=True)
edu_df['MainBranch'].replace(np.nan,'No answer',regex=True,inplace=True)
edu_df['Gender'].replace(np.nan,'No answer',regex=True,inplace=True)

# Replace values
mainbranch_dict = {
    "I am a developer by profession": "Professional developer",
    "I am learning to code": "Learn to code",
    "I am not primarily a developer, but I write code sometimes as part of my work"  : "Ability to code",
    "I code primarily as a hobby": "Ability to code",
    "I used to be a developer by profession, but no longer am": "Ability to code",
    "None of these" : "None",
    "No answer":"None" }

edu_dict = {
    "Something else": "No answer",
    "Associate degree (A.A., A.S., etc.)": "Below bachelor",
    "Bachelor’s degree (B.A., B.S., B.Eng., etc.)": "Bachelor",
    "Master’s degree (M.A., M.S., M.Eng., MBA, etc.)": "Master",
    "Other doctoral degree (Ph.D., Ed.D., etc.)": "Doctoral",
    "Professional degree (JD, MD, etc.)": "Professional degree",
    "Secondary school (e.g. American high school, German Realschule or Gymnasium, etc.)": "Below bachelor",
    "Some college/university study without earning a degree": "Below bachelor",
    "Primary/elementary school": "Below bachelor",
}

edu_df = edu_df.replace({"EdLevel": edu_dict})
edu_df = edu_df.replace({"MainBranch": mainbranch_dict})
```

```python
# Cross table of MainBranch and EdLevel
edu_table = pd.crosstab(edu_df['MainBranch'], edu_df['EdLevel'],
rownames=['MainBranch'], colnames=['EdLevel'])
branch_order = ['Professional developer', 'Learn to code', 'Ability to code']
edu_order = ['Below bachelor', 'Bachelor', 'Master', 'Doctoral', 'Professional degree']

edu_table = edu_table.reindex(branch_order)
edu_table = edu_table.reindex(edu_order, axis=1)

# Style table
edu_table.style.background_gradient(cmap='Blues').set_caption('Split developer profile based on Education Level and Main Branch')
```

```python
# Group by MainBranch and sum, set index to MainBranch
branch_learn_group_df = learn_2_code_split_df.groupby(['MainBranch'])['School','Bootcamp','Online learning','Books','Social learning','Hackathons'].sum()
branch_learn_group_df
```

```python
# Set index to MainBranch
branch_learn_group_df.reset_index(inplace=True)
branch_learn_group_df.set_index('MainBranch', inplace=True)

# Flatten the dataframe
branch_learn_group_df = branch_learn_group_df.stack().reset_index()
```

```python
# Rename columns and sort values, remove rows with 0 and create crosstab for visualization
branch_learn_group_df.rename(columns={'level_1':'LearnCode', 0:'Count'}, inplace=True)
branch_learn_group_df.sort_values(by='Count', ascending=False, inplace=True)
branch_learn_group_df = branch_learn_group_df.loc[branch_learn_group_df['Count'] > 0]
branch_learn_group_ct = pd.crosstab(branch_learn_group_df['LearnCode'], branch_learn_group_df['MainBranch'], values=branch_learn_group_df['Count'], aggfunc='sum', normalize='columns').round(4)*100
```

```python
fig = px.bar(branch_learn_group_ct.transpose(),
    title = "Percentage of learning code methods for each Main Branch classification",
    labels = {"value": "Percentage(%)", "variable": "Main Branch"},
)
fig.update_layout(barmode='group')
fig.show(  )
```
